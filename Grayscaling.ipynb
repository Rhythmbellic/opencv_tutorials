```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[1], image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```
