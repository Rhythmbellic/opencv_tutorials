```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def imshow(title="",image=None,size=10):
  w,h=image.shape[1],image.shape[0]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
blank_image2=np.zeros((512,512),np.uint8)
```

```python
imshow('blan',blank_image)
imshow('blank',blank_image2)
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
cv2.line(blank_image,(0,0),(511,511),(0,255,255),5)
imshow("line",blank_image)
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
cv2.rectangle(blank_image,(100,100),(300,250),(255,255,255),5)
imshow("rectangle",blank_image)
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
cv2.circle(blank_image,(350,350),100,(0,255,0),-1)
imshow("line",blank_image)
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
pts=np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)
pts=pts.reshape((-1,1,2))
cv2.polylines(blank_image,[pts],True,(0,0,255),3)
imshow("bla",blank_image)
```

```python
blank_image=np.zeros((512,512,3),np.uint8)
string2='roodie loves you all'
pts=pts.reshape((-1,1,2))
cv2.putText(blank_image,string2,(155,290),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
imshow("bla",blank_image)
```
